```python
#loading the dataset
df = pd.read_excel('D:/DA/dataset/online+retail/Online Retail.xlsx')
```

```python
dataset = df[df.Country == "United Kingdom"]
```

```python
#Dropping the rows with missing data
dataset = dataset.dropna(axis=0)
```

```python
#Dropping the rows with cancelled transactions
dataset = dataset[pd.to_numeric(dataset.InvoiceNo, errors='coerce').notnull()]
dataset.InvoiceNo = dataset.InvoiceNo.astype('int32')
```

```python
dataset['Description'].value_counts()
```
